```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(r'C:\Users\GrowHouse NYC\PythonProject\data\raw\dportal_list_activities (JAM).csv')


# Create a new DataFrame with 'title' and 'funding' columns
new_df = pd.DataFrame({+-
    'title': df['title'],  # Extract data from column 'B' and assign it to 'title'
    'funding': df['total-spend']  # Extract data from column 'E' and assign it to 'funding'
})

# Sort the DataFrame based on the 'funding' column in descending order
sorted_df = new_df.sort_values(by='funding', ascending=False)

# Print the sorted DataFrame in a neat format
print("Sorted Data (Descending by Funding):")
print(sorted_df.to_string(index=False))
```
